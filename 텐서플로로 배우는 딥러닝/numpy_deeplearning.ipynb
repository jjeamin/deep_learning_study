# 정보
놀라움의 정도

$h(x) = -logP(x)$

# Entropy
정보 이론에서 정보의 불확실함의 정도

- 자주 발생하는 사건은 낮은 정보량을 가진다. 발생이 보장된 사건은 그 내용에 상관없이 전혀 정보가 없다는 걸 뜻한다.
- 덜 자주 발생하는 사건은 더 높은 정보량을 가진다.

$entropy = E(-logP(x)) = \sum P(x)(-logP(x))$

# Cross Entropy
두 확률분포가 교차로 곱해진다는 의미

$entropy = E(-logP(x)) [minimize]$

실제 값(P)과 예측값(Q)의 차이

$\sum P(x)*-log(Q(x))$

비슷할수록 값은 작아지고 다를수록 값은 커진다

```python
import numpy as np

x_train = []
t_train = []
x_test = []
t_test = []

mu, sigma = 0, 0.1

for i in range(0,3000):
    data = np.array([0,0,0])- np.random.normal(mu, sigma, 3)
    x_train.append(data)
    t_train.append([1,0,0,0,0,0,0,0])
    
    data = np.array([0,0,1])- np.random.normal(mu, sigma, 3)
    x_train.append(data)
    t_train.append([0,1,0,0,0,0,0,0])      
    
    data = np.array([0,1,0])- np.random.normal(mu, sigma, 3)
    x_train.append(data)
    t_train.append([0,0,1,0,0,0,0,0])
      
    data = np.array([0,1,1])- np.random.normal(mu, sigma, 3)
    x_train.append(data)
    t_train.append([0,0,0,1,0,0,0,0])
    
    data = np.array([1,0,0])- np.random.normal(mu, sigma, 3)
    x_train.append(data)
    t_train.append([0,0,0,0,1,0,0,0])
     
    data = np.array([1,0,1])- np.random.normal(mu, sigma, 3)
    x_train.append(data)
    t_train.append([0,0,0,0,0,1,0,0])
    
    data = np.array([1,1,0])- np.random.normal(mu, sigma, 3)
    x_train.append(data)
    t_train.append([0,0,0,0,0,0,1,0])
    
    data = np.array([1,1,1])- np.random.normal(mu, sigma, 3)
    x_train.append(data)
    t_train.append([0,0,0,0,0,0,0,1])
    
x_train = np.array(x_train)
t_train = np.array(t_train)

print(x_train.shape)
print(t_train.shape)

for i in range(0,2000):
    data = np.array([0,0,0])- np.random.normal(mu, sigma, 3)
    x_test.append(data)
    t_test.append([1,0,0,0,0,0,0,0])
    
    data = np.array([0,0,1])- np.random.normal(mu, sigma, 3)
    x_test.append(data)
    t_test.append([0,1,0,0,0,0,0,0])
    
    data = np.array([0,1,0])- np.random.normal(mu, sigma, 3)
    x_test.append(data)
    t_test.append([0,0,1,0,0,0,0,0])
    
    data = np.array([0,1,1])- np.random.normal(mu, sigma, 3)
    x_test.append(data)
    t_test.append([0,0,0,1,0,0,0,0])
    
    data = np.array([1,0,0])- np.random.normal(mu, sigma, 3)
    x_test.append(data)
    t_test.append([0,0,0,0,1,0,0,0])
    
    data = np.array([1,0,1])- np.random.normal(mu, sigma, 3)
    x_test.append(data)
    t_test.append([0,0,0,0,0,1,0,0])
    
    data = np.array([1,1,0])- np.random.normal(mu, sigma, 3)
    x_test.append(data)
    t_test.append([0,0,0,0,0,0,1,0])
    
    data = np.array([1,1,1])- np.random.normal(mu, sigma, 3)
    x_test.append(data)
    t_test.append([0,0,0,0,0,0,0,1])
    
x_test = np.array(x_test)
t_test = np.array(t_test)
```

```python
def relu(x):
    return np.maximum(0,x)

def relu_grad(x):
    mask = (x <= 0)
    x[mask] = 0
    return x

def softmax(z):
    assert len(z.shape) == 2
    s = np.max(z, axis=1)
    s = s[:, np.newaxis]
    e_x = np.exp(z - s)
    div = np.sum(e_x, axis=1)
    div = div[:, np.newaxis]
    return e_x / div

def cross_entropy_error(y, t):
    a = 1e-7
    
    return np.mean(-np.sum(y * np.log(t + a)))
```

```python
#하이퍼 파라미터 = 값을 정해줘야하는 것
input_node = 3
output_node = 8
epoch = 100
train_size = x_train.shape[0] #학습 사이즈
batch_size = 128 #배치사이즈 = 한번에 학습시키는 회수 128개씩
lr = 0.01 #학습률
train_loss_list = []

W1 = np.random.rand(input_node ,12) 
W2 = np.random.rand(12 ,output_node)

for j in range(50000):
    batch_mask = np.random.choice(train_size,batch_size)
    train_data = x_train[batch_mask]
    label = t_train[batch_mask]

    layer1 = np.dot(train_data, W1) 
    func1 = relu(layer1)
    out = np.dot(func1, W2)

    pred = softmax(out)
    loss = cross_entropy_error(pred,label)

    if i % 100:
        print(loss / batch_size)
        
    #backpropagation
    dout = (pred - label) / batch_size

    new_W2 = W2 + (-lr)*np.dot(func1.T, dout)
    
    dout = np.dot(dout, W2.T) * relu_grad(layer1)
    
    new_W1 = W1 + (-lr)*np.dot(train_data.T, dout)

    W2 = new_W2
    W1 = new_W1
```

```python
layer1 = np.dot(x_test, W1) 
func1 = relu(layer1)
out = np.dot(func1, W2)

pred = softmax(out)

print(cross_entropy_error(pred,t_test) / len(t_test))

print(np.sum(np.argmax(pred,axis=1) == np.argmax(t_test,axis=1))/len(t_test) * 100)
```
