```python
x_train = []    
t_train = []    
x_test = []       
t_test = []       

for i in range(0,3000):
    data = np.array([0,0,0]) + np.random.normal(0, 0.1, 3)
    x_train.append(data)
    t_train.append([1,0,0,0])

    data = np.array([0,0,1]) + np.random.normal(0, 0.1, 3)
    x_train.append(data)
    t_train.append([0,1,0,0])      

    data = np.array([0,1,0]) + np.random.normal(0, 0.1, 3)
    x_train.append(data)
    t_train.append([0,0,1,0])

    data = np.array([0,1,1]) + np.random.normal(0, 0.1, 3)
    x_train.append(data)
    t_train.append([0,0,0,1])
    
    data = np.array([1,0,0]) + np.random.normal(0, 0.1, 3)
    x_train.append(data)
    t_train.append([0,0,0,1])

    data = np.array([1,0,1]) + np.random.normal(0, 0.1, 3)
    x_train.append(data)
    t_train.append([0,0,1,0])

    data = np.array([1,1,0]) + np.random.normal(0, 0.1, 3)
    x_train.append(data)
    t_train.append([0,1,0,0])

    data = np.array([1,1,1]) + np.random.normal(0, 0.1, 3)
    x_train.append(data)
    t_train.append([1,0,0,0])

for i in range(0,1000):
    data = np.array([0,0,0]) + np.random.normal(0, 0.1, 3)
    x_test.append(data)
    t_test.append([1,0,0,0])

    data = np.array([0,0,1]) + np.random.normal(0, 0.1, 3)
    x_test.append(data)
    t_test.append([0,1,0,0])      

    data = np.array([0,1,0]) + np.random.normal(0, 0.1, 3)
    x_test.append(data)
    t_test.append([0,0,1,0])

    data = np.array([0,1,1]) + np.random.normal(0, 0.1, 3)
    x_test.append(data)
    t_test.append([0,0,0,1])
    
    data = np.array([1,0,0]) + np.random.normal(0, 0.1, 3)
    x_test.append(data)
    t_test.append([0,0,0,1])

    data = np.array([1,0,1]) + np.random.normal(0, 0.1, 3)
    x_test.append(data)
    t_test.append([0,0,1,0])

    data = np.array([1,1,0]) + np.random.normal(0, 0.1, 3)
    x_test.append(data)
    t_test.append([0,1,0,0])

    data = np.array([1,1,1]) + np.random.normal(0, 0.1, 3)
    x_test.append(data)
    t_test.append([1,0,0,0])

x_train = np.array(x_train)
t_train = np.array(t_train)
x_test = np.array(x_test)
t_test = np.array(t_test)

print(x_train.shape)
print(t_train.shape)
print(x_test.shape)
print(t_test.shape)
```

```python
kernel_initializer = tf.keras.initializers.glorot_normal(seed=None)

model = tf.keras.models.Sequential([
  tf.keras.Input(shape=3),
  tf.keras.layers.Dense(5, activation='relu'
                        ,kernel_initializer=kernel_initializer),
  tf.keras.layers.Dense(4, activation='softmax'
                        ,kernel_initializer=kernel_initializer)
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
```

```python
model.fit(x_train, t_train, epochs=5)
```

# metrics


      실제 True  |  False
    예측  ==================
    True  |  TP   |   FP   |
          ==================
    False |  FN   |   TN   |
          ==================


## recall
- 재현율
- 정답으로 예측한 값의 개수 / 실제 정답의 개수


$$\frac{TP}{TP + FN}$$


## precision
- 정밀도
- 정답으로 예측한 값의 개수 / 예측한 정답의 개수


$$\frac{TP}{TP + FP}$$


## f1 score
- recall 과 precision의 조화 평균

$$2*\frac{Precision*Recall}{Precision+Recall}$$


## Accuracy
- 틀린것을 틀렸다고 하는것도 정답으로하는 metrics

$$\frac{TP+TN}{TP+TN+FP+FN}$$


## roc
- FPR(거짓 양성 비율)과 TPR(진짜 양성 비율)의 곡선


## auc
- roc 곡선 아래의 면적
- 1에 가까울수록 좋다

```python
from tensorflow.keras import backend as K

def recall(y_target, y_pred):
    # clip(t, clip_value_min, clip_value_max) : clip_value_min~clip_value_max 이외 가장자리를 깎아 낸다
    # round : 반올림한다
    y_target_yn = K.round(K.clip(y_target, 0, 1)) # 실제값을 0(Negative) 또는 1(Positive)로 설정한다
    y_pred_yn = K.round(K.clip(y_pred, 0, 1)) # 예측값을 0(Negative) 또는 1(Positive)로 설정한다

    # True Positive는 실제 값과 예측 값이 모두 1(Positive)인 경우이다
    count_true_positive = K.sum(y_target_yn * y_pred_yn) 

    # (True Positive + False Negative) = 실제 값이 1(Positive) 전체
    count_true_positive_false_negative = K.sum(y_target_yn)

    # Recall =  (True Positive) / (True Positive + False Negative)
    # K.epsilon()는 'divide by zero error' 예방차원에서 작은 수를 더한다
    recall = count_true_positive / (count_true_positive_false_negative + K.epsilon())

    # return a single tensor value
    return recall


def precision(y_target, y_pred):
    # clip(t, clip_value_min, clip_value_max) : clip_value_min~clip_value_max 이외 가장자리를 깎아 낸다
    # round : 반올림한다
    y_pred_yn = K.round(K.clip(y_pred, 0, 1)) # 예측값을 0(Negative) 또는 1(Positive)로 설정한다
    y_target_yn = K.round(K.clip(y_target, 0, 1)) # 실제값을 0(Negative) 또는 1(Positive)로 설정한다

    # True Positive는 실제 값과 예측 값이 모두 1(Positive)인 경우이다
    count_true_positive = K.sum(y_target_yn * y_pred_yn) 

    # (True Positive + False Positive) = 예측 값이 1(Positive) 전체
    count_true_positive_false_positive = K.sum(y_pred_yn)

    # Precision = (True Positive) / (True Positive + False Positive)
    # K.epsilon()는 'divide by zero error' 예방차원에서 작은 수를 더한다
    precision = count_true_positive / (count_true_positive_false_positive + K.epsilon())

    # return a single tensor value
    return precision


def f1score(y_target, y_pred):
    _recall = recall(y_target, y_pred)
    _precision = precision(y_target, y_pred)
    # K.epsilon()는 'divide by zero error' 예방차원에서 작은 수를 더한다
    _f1score = ( 2 * _recall * _precision) / (_recall + _precision+ K.epsilon())
    
    # return a single tensor value
    return _f1score

auc = tf.keras.metrics.AUC(num_thresholds=3)
```

```python
kernel_initializer = tf.keras.initializers.glorot_normal(seed=None)

model = tf.keras.models.Sequential([
  tf.keras.Input(shape=3),
  tf.keras.layers.Dense(5, activation='relu'
                        ,kernel_initializer=kernel_initializer),
  tf.keras.layers.Dense(4, activation='softmax'
                        ,kernel_initializer=kernel_initializer)
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['categorical_accuracy',recall,precision,f1score,auc])
```

```python
model.fit(x_train, t_train, epochs=5)
```

```python
result = model.evaluate(x_test,t_test)
```
