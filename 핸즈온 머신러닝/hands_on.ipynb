# 머신러닝 프로젝트

## 큰 그림을 본다.
1. 목표를 비즈니스 용어로 정의한다.


2. 이 솔루션은 어떻게 사용될 것인가?


3. (만약 있다면)현재 솔루션이나 차선책은 무엇인가?


4. 어떤 문제라고 정의할 수 있나(지도/비지도, 온라인/오프라인)?


5. 성능을 어떻게 측정해야 하나?


6. 성능 지표가 비즈니스 목표에 연결되어 있나?


7. 비즈니스 목표에 도달하기 위해 필요한 최소한의 성능은 얼마인가?


8. 비슷한 문제가 있나? 이전의 방식이나 도구를 재사용할 수 있나?


9. 해당 분야의 전문가가 있나?


10. 수동으로 문제를 해결하는 방법은 무엇인가?


11. 여러분이 세운 가정을 나열한다.


12. 가능하면 가정을 검증한다.

## 데이터를 수집한다.
1. 필요한 데이터와 양을 나열한다.


2. 데이터를 얻을 수 있는 곳을 찾아 기록한다.


3. 얼마나 많은 공간이 필요한지 확인한다.


4. 법률상의 의무가 있는지 확인하고 필요하다면 인가를 받는다.


5. 접근 권한을 획득한다.


6. 작업 환경을 만든다.


7. 데이터를 수집한다.


8. 데이터를 조작하기 편한 형태로 변환한다.(전처리)


9. 민감한 정보가 삭제되었거나 보호되었는지 검증한다.


10. 데이터의 크기와 타입(시계열, 표본, 지리정보 등)을 확인한다.


11. 테스트 세트를 샘플링하여 따로 떼어놓고 절대 들여다 보지 않는다.

## 데이터를 탐색한다.
1. 데이터 탐색을 위해 복사본을 생성한다.


2. 데이터 탐색 결과를 저장하기 위해 주피터 노트북을 만든다.


3. 각 특성의 특징을 조사한다.
    *이름 타입 비율 잡음 유용여부 분포형태*  
    
    
4. 지도 학습 작업이라면 타깃 속성을 구분한다.


5. 데이터를 시각화 한다.


6. 특성 간의 상관관계를 조사한다.


7. 수동으로 문제를 해결할 수 있는 방법을 찾아본다.


8. 적용이 가능한 변환을 찾는다.


9. 추가로 유용한 데이터를 찾는다.(있으면 "데이터를 수집한다"로 돌아간다.)


10. 조사한 것을 기록한다.

## 데이터를 준비한다.
```
데이터 복사본으로 작업한다.

적용한 모든 데이터 변환은 함수로 만든다.
```

1. 데이터 정제
    - 이상치를 수정/삭제
    - 누락된 값을 채우거나 행/열 제거
    
</br>

2. 특성 선택
    - 유용하지 않은 정보를 가진 특성 제거
   
</br>
   
3. 적절한 특성 공학
    - 연속 특성 이산화하기
    - 특성 분해하기( 범주형, 날짜/시간 etc. )
    - 가능한 특성 변환 추가하기( $log(x),sqrt(x)$ etc. )
    - 특성을 조합해 가능성 있는 새로운 특성 만들기


4. 특성 스케일 조정: 표준화, 정규화

## 가능성 있는 몇 개의 모델을 고른다.
1. 여러 종류의 모델을 기본 매개변수를 사용해 신속하게 많이 훈련 시켜 본다. 
    - 선형 모델
    - 나이브 베이지
    - SVM
    - 랜덤 포레스트
    - 신경망 
    - etc.
   
   
2. 성능을 측정/비교
    - 각 모델에서 N-겹 교차 검증을 사용해 N개 폴드(집합)의 성능에 대한 평균과 표준편차를 계산
    
    
3. 각 알고리즘에서 가장 두드러진 변수를 분석한다.


4. 모델이 만드는 에러의 종류를 분석한다.
    - 이 에러를 피하기 위해 사람이 사용하는 데이터는 무엇인가요??
    
    
5. 간단한 특성 선택과 특성 공학 단계를 거친다.


6. 이전 다섯 단계를 한 번이나 두 번 빠르게 반복한다.


7. 다른 종류의 에러를 만드는 모델을 중심으로 가장 가능성이 높은 모델을 3~5개 정도 추린다.

## 시스템을 세밀하게 튜닝한다.

1. 교차 검증을 사용해 하이퍼파라미터를 정밀 튜닝 한다.
    - 탐색할 하이퍼파라미터 값이 매우 적지 않다면 랜덤 서치를 사용하고, 훈련시간이 오래 걸리면 베이지안 최적화 방법을 사용하는 것이 좋다.(**가우시안 프로세스 사전확률**)


2. 앙상블 방법을 시도해 본다.


3. 최종 모델에 확신이 선 후 일반화 오차를 추정하기 위해 테스트 세트에서 성능을 측정한다.

    - 먼저 큰 그림을 부각시킨다.


3. 이 솔루션이 어떻게 비즈니스의 목표를 달성하는지 설명하세요

5. 멋진 그래프나 기억하기 쉬운 문장으로 핵심 내용을 전달하세요
